# **روش 1 : محاسبه اندیکاتور میانگین متحرک نمایی  در بازار سهام**

```python
# Define the parameters
symbol = 'ETH-USD'
import yfinance as yf
start_date = '2022-06-01'
end_date = '2023-06-01'

short_window = 10
long_window = 30




data = yf.download(symbol, start=start_date, end=end_date)
print(data)
data['ema_short'] = ta.trend.ema_indicator(data['Close'], window=short_window)
data['ema_long'] = ta.trend.ema_indicator(data['Close'], window=long_window)

import matplotlib.pyplot as plt

plt.plot(data['ema_short'],label='ema_short')

plt.plot(data['ema_long'],label='ema_long')

plt.plot(data['Close'],label='Close')

plt.legend()
plt.title('Exponential Moving Average(EMA)')
plt.show()

```

# **روش 2 : محاسبه اندیکاتور میانگین متحرک نمایی  در بازار سهام**

```python
# Define the parameters
symbol = 'ETH-USD'
import yfinance as yf
start_date = '2022-06-01'
end_date = '2023-06-01'

short_window = 10
long_window = 30




data = yf.download(symbol, start=start_date, end=end_date)
print(data)

data['ema_short']=data['Close'].ewm(span=short_window).mean()
data['ema_long']=data['Close'].ewm(span=long_window).mean()


import matplotlib.pyplot as plt

plt.plot(data['ema_short'],label='ema_short')

plt.plot(data['ema_long'],label='ema_long')

plt.plot(data['Close'],label='Close')

plt.legend()
plt.title('Exponential Moving Average(EMA)')
plt.show()

```
